```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sqlalchemy import create_engine
from sqlalchemy.engine.url import URL
import config
from config import CONNSTRING
from sklearn.model_selection import train_test_split
```

```python
#read the data from SQL
engine= create_engine(CONNSTRING)

query = "SELECT * FROM customer_data"
df = pd.read_sql(query, engine)
        
```

```python

null_count = df['age'].isnull().sum()
print(f'there are {null_count} null values')
```

```python
customer_segmentation = df[['age', 'gender', 'category', 'Purchase Amount (USD)', 'location', 'season', 'Previous Purchases', 'Frequency of Purchases', 'Promo Code Used','Payment Method']].copy()
customer_segmentation_encoded = pd.get_dummies(customer_segmentation, columns=['gender','category','location','season','Frequency of Purchases','Promo Code Used','Payment Method'])
customer_segmentation_encoded.drop(['Promo Code Used_No','gender_Male'], axis=1, inplace=True)

```

```python
customer_segmentation_encoded.head()
```

```python
X = customer_segmentation_encoded.drop('age', axis=1)
y = customer_segmentation_encoded['age']
```

```python
#divde train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.32, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

#create mode
model = RandomForestClassifier(random_state=42)

#fit model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

#make predictions
y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
